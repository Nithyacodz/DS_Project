```python
# Import the numpy and pandas packages

import numpy as np
import pandas as pd
```

```python
# importing the Companies text file and listing the top few row. The data is delimited by Tab and encoding is latin

companies = pd.read_csv("companies.txt", sep='\t', encoding='ISO-8859-1')
companies.head()
```

```python
# Removing special characters in a pandas dataframe

companies.permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
companies.head()
```

```python
# permalink field has various cases for same name - upper and lower, hence converting all to lower

companies['permalink'] = companies['permalink'].str.lower() 
companies.head()
```

```python
# importing the rounds2 CSV file and listing the top few row. 

rounds2 = pd.read_csv("rounds2.csv", encoding='ISO-8859-1')
rounds2.head()
```

```python
# Removing special characters in a pandas dataframe

rounds2.company_permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
rounds2.head()
```

```python
# permalink field has various cases for same name - upper and lower, hence converting all to lower

rounds2['company_permalink'] = rounds2['company_permalink'].str.lower() 
rounds2.head()
```

```python
# unique field in companies dataframe is name field. Hence identifying the unique names and saving them in a new dataframe called "unique_companies"
# inspecting the shape of unique_companies dataframe to obtain the number of rows. 
# "names" field could also be used but since we intend to use permalink in Companies as it is common to company_permalink from rounds2

unique_companies = pd.DataFrame(companies.permalink.unique())
unique_companies.shape

```

```python
# unique field in rounds2 dataframe is company_permalink field. Hence identifying the unique names and saving them in a new dataframe called "unique_rounds"
# inspecting the shape of unique_rounds2 dataframe to obtain the number of rows

unique_rounds2 = pd.DataFrame(rounds2.company_permalink.unique())

unique_rounds2.shape

```

```python



#In the companies data frame, which column can be used as the unique key for each company? Write the name of the column.

# Answer - company_permalink
```

```python
# merging companies and round2 dataframes to create a new dataframe called masterframe

masterframe = pd.merge(rounds2, companies, left_on=['company_permalink'], right_on = ['permalink'], how='left')
masterframe.shape
```

```python
#to find out the number of cases where company names exists in rounds2 dataframe and not in companies2. 
# the sum of NUll values in column permalink which is from companies dataframe.

masterframe['raised_amount_usd'].isnull().sum()

```

```python
# inspecting the first few rows of masterframe

masterframe.head()
```

```python
# Creating a new dataframe called fundingtype_masterframe after all "NaN" values in column "raised_amount_usd" from masterframe

masterframe = masterframe[~np.isnan(masterframe['raised_amount_usd'])]


```

```python
# converting the USD column to integer

masterframe.raised_amount_usd = masterframe.raised_amount_usd.astype(np.int64)

# inspecting the first few rows of masterframe

masterframe.head()
```

```python
# grouping the data in masterframe by "funding round type"


fundingtype_masterframe = masterframe.groupby('funding_round_type')


# finding the mean of "raised amount in USD grouped by "funding round type" and rounding the result 

fundingtype_masterframe = round(pd.DataFrame(fundingtype_masterframe.raised_amount_usd.mean()),0)

# displaying the results of the dataframe

fundingtype_masterframe = fundingtype_masterframe.sort_values(by = ['raised_amount_usd'], ascending = False)
fundingtype_masterframe
```

```python
# finding the count of various investment types.


fundingtype_masterframe_count = masterframe.groupby('funding_round_type')

fundingtype_masterframe_count = round(pd.DataFrame(fundingtype_masterframe_count.raised_amount_usd.count()),0)

fundingtype_masterframe_count
```

```python
# identifying investment type between 5 to 15 million USD per investment round

fundingtype_masterframe['raised_amount_usd'].between(5000000, 15000000, inclusive=True)

# venture is most suitable
```

```python
# creating a seperate dataframe containing only funding type as venture

venture_masterframe = masterframe.loc[(masterframe.funding_round_type == "venture"), :]
venture_masterframe.head()
```

```python
# grouping by Country Code in dataframe called top9_alltypes . (for all investment types)
top9_alltypes = masterframe.groupby('country_code')


# aggregating and displaying based raised amount in USD
top9_alltypes = pd.DataFrame(top9_alltypes.raised_amount_usd.sum())


# the data type of the value field is float converting to integer
top9_alltypes.raised_amount_usd = top9_alltypes.raised_amount_usd.astype(np.int64)


# sorting in descending order based on raised amount in USD
top9_alltypes = top9_alltypes.sort_values(by = ['raised_amount_usd'], ascending = False)


# extracting the top9_alltypes  and saving in dataframe top9_alltypes 
top9_alltypes = top9_alltypes.iloc[0:9,:]

# displaing contents of dataframe
top9_alltypes
```

```python
# creating a seperate dataframe containing only funding type as venture called Top9

top9 = masterframe.loc[(masterframe.funding_round_type == "venture"), :]

# grouping by Country Code in dataframe called top9 
top9 = top9.groupby('country_code')


# aggregating and displaying based raised amount in USD
top9 = pd.DataFrame(top9.raised_amount_usd.sum())


# the data type of the value field is float converting to integer
top9.raised_amount_usd = top9.raised_amount_usd.astype(np.int64)


# sorting in descending order based on raised amount in USD
top9 = top9.sort_values(by = ['raised_amount_usd'], ascending = False)



# extracting the top 9 and saving in dataframe top9
top9 = top9.iloc[0:9,:]

# displaing contents of dataframe, TOP 9 Countries for Venture  based on investment raised
top9
```

```python
# The business rule that the first string before the vertical bar will be considered the primary sector. 


# creating a new dataframe called primarysector to save the split of category list column. Splitting is based on pipe 

primarysector = venture_selectcountry_masterframe["category_list"].str.split("|", expand = True)


# update venture_selectcountry_masterframe dataframe with new column gprimarysector based on 1st column of primarysector dataframe
venture_selectcountry_masterframe["primarysector"]= primarysector[0] 

# Inspecting the dataframe to see if the new column is now added
venture_selectcountry_masterframe.info()

```

```python
# Loading the mapping csv file

mapping = pd.read_csv("mapping.csv")
mapping.head()
```

```python
mapping.info()
```

```python
# reshaping the dataframe

mapping = pd.melt(mapping, id_vars=["category_list"], var_name="Sector", value_name="existing")

# inspecting the dataframe after reshaping
mapping.head()
```

```python
# since a value '0' in existing column indicates there is no mapping between sector and category list. Hence deleting all rows with value in existing as 0

mapping.drop(mapping[mapping.existing == 0].index, inplace=True)

# inspecting the new dataframe
mapping.head()
```

```python
# inspecting the contents after reshaping and removing the unecessary rows

mapping.info()
```

```python

mapping.tail(100)
# on inspection noticed a few category list have "0" instead of "na" such as A0lytics, See index (5777,6020,6038,6054, etc)
```

```python
# finding category list containing '0'

mapping['category_list'].str.contains('0')

# on inspection you will find few which are True and needs correction
```

```python

# as we found many instances where instead of 'na' it contained 0, we are finding 0 in category_list string and replacing with na. Eg index numbers - 20, 664
#print(mapping.category_list.str.replace('0','na'))

mapping['category_list']=mapping.category_list.replace({"0": "na"},regex=True)

# inspecting the result. See index (5777,6020,6038,6054, etc)
mapping.tail(100)
```

```python
# delete the column existing as it is no longer required

mapping.pop('existing')

# inspecting the updated dataframe
mapping.head()
```

```python

# New DataFrame called D1 from earier dataframe for investments betweeen USD 5 Million and 15 Million in USA.
# No filter considered for Investment type "Venture" as the sect_country_master was already filtered to contain investment type venture only

D1 = sect_country_master[(sect_country_master['raised_amount_usd'].between(5000000, 15000000, inclusive=True)) & (sect_country_master['country_code'] == "USA")]

# Inspecting the contents of new database
D1.head()
```

```python
# creating a function for computing the sum and count.

def sum_col(df, col, new_col):
    df[new_col] = df[col].sum()
    return df

def count_col(df, col, new_col):
    df[new_col] = df[col].count()
    return df

# adding sector wise sum of investment and count

D1 = D1.groupby("Sector").apply(sum_col, 'raised_amount_usd', 'sector_investment_sum')
D1 = D1.groupby("Sector").apply(count_col, 'raised_amount_usd', 'sector_investment_count')
D1.head()
```

```python
# Total Investments in the country


D1['raised_amount_usd'].count()

# The Total Investments is 12098
```

```python
# Total investment in country USA

D1['raised_amount_usd'].sum()

# The Total Investments is $108083057956
```

```python
# grouping the data in masterframe by "Sector" to find the top counts of investments by Sector


D1_sector = D1.groupby('Sector').count()
D1_sector.sort_values(by = (['raised_amount_usd']), ascending = False)


# Top three sectors investment count wise - 1. Others, 2. Social, Finance, Analytics, Advertising and 
# 3. Cleantech / Semiconductors

```

```python
# Filter for all investments in Others Sector being the Top Sector

D1_top = D1.loc[(D1.Sector == "Others"), :]

# Group by name (Company Name) and sort based on raised amount to sort companies based on amount received

D1_top = D1_top.groupby('name').sum()
D1_top = D1_top.sort_values(by = (['raised_amount_usd']), ascending = False)
D1_top['raised_amount_usd'].head(3)

# Virtustream received the highest investment in Others Sector
```

```python
# Filter for all investments in Social, Finance, Analytics, Advertising Sector being the Second Sector


D1_top_2 = D1.loc[(D1.Sector == "Social, Finance, Analytics, Advertising"), :]

# Group by name (Company Name) and sort based on raised amount to sort companies based on amount received

D1_top_2 = D1_top_2.groupby('name').sum()
D1_top_2 = D1_top_2.sort_values(by = (['raised_amount_usd']), ascending = False)
D1_top_2['raised_amount_usd'].head(3)

# SST Inc. (Formerly ShotSpotter) received the second highest investment in Social, Finance, Analytics, Advertising
```
